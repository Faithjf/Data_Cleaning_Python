- *it splits the value into feet and inches*
- *converts them to centimeters using the conversion factors of 30.48 and 2.54 respectively.*

```python
# Now all the values will be in cm, even though i remove the cm.

#df['Height'] = df['Height'].apply(lambda x: int(x.split("'")[0]) * 30.48 + 
                                       #int(x.split("'")[1].replace('"', '')) * 2.54 if "'" in x else x)
    
df['Height'] = df['Height'].apply(lambda x: 
                                  int(x.split("'")[0]) * 30.48 + int(x.split("'")[1].replace('"', '')) * 
                                  2.54 if isinstance(x, str) and "'" in x else x)

df['Height']= df['Height'].replace('cm', '', regex=True)

df.Height.unique()
```

### Inspecting for outliers in our data

```python
# inspecting for outliers in our data

def outlier_lims(col):
    q3,q1 = np.percentile(col, [75,25])
    iqr = q3-q1
    upper_lim = q3 + 1.5*iqr
    lower_lim = q1 - 1.5*iqr
    return upper_lim, lower_lim

for col in continuous_vars:
    print("--------------------------------------------------")
    print("Column:", col)
    
    UL,LL = outlier_lims(df_cleaned[col])
    print("Upper Limit =", UL)
    print("Lower Limit =", LL)
    
    total_outliers = len(df_cleaned.loc[df_cleaned[col]<LL,col]) + len(df_cleaned.loc[df_cleaned[col]>UL,col])
    percent = (total_outliers / len(df_cleaned.index) )*100
    
    print("Percentage of Outliers=", percent)
    print("----------------------------------------------")
```
